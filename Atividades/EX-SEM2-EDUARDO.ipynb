#**NLTK**

# Tokenização
Aqui será feito o processo de quebra de cada frase em palavras.

```python
nltk.download('punkt')
```

```python
# Tokenização NLTK 
textTokenNLTK = []
for word in texts:
    textTokenNLTK.append(nltk.word_tokenize(word))
```

```python
# Resultado da tokenização
for i in range(len(textTokenNLTK)):
    print("Frase "+str(i+1)+ " tokenização : " + str(textTokenNLTK[i]))
```

# Remoção das Stop Words
Removendo as palavras mais comuns do texto que acabam por não gerar valor 

```python
# Baixando a biblioteca com as palavras em português 
nltk.download('stopwords')
stopWords = nltk.corpus.stopwords.words('portuguese')
```

```python
# Stop Word NLTK

wordsFilteredNLTK = []
# Como é uma array dentro de uma array é preciso realizar dois loops
for elementArray in textTokenNLTK:
    filteredPhrases = []
    for element in elementArray:
        if element not in stopWords:
            wordsFilteredNLTK.append(element)
            filtered_phrase =''.join(element)
            filteredPhrases.append(filtered_phrase)



```

```python
# Stop Word NLTK
# Criando um index com as listas para usar com a array dupla
resultStopNLTK = [[]for _ in range(len(textTokenNLTK))]

for x, phrase in enumerate (textTokenNLTK):
    for word in phrase:
        if word not in stopWords:
            resultStopNLTK[x].append(word)
```

```python
# Resultado da remoção das stop words
for i in range(len(resultStopNLTK)):
    print("Frase "+str(i+1)+ " Stop Words : " + str(resultStopNLTK[i]))
```

# Stemming 

Stemming é o processo de colocar as palavras no seu formato base, assim simplificando as coisas para o modelo. 


```python
nltk.download('rslp')
```

```python
stemmer = nltk.stem.RSLPStemmer()
```

```python
# Stemming NLTK
# Criando um index com as listas para usar com a array dupla
resultStemming = [[]for _ in range(len(resultStopNLTK))]

for x, phrase in enumerate (resultStopNLTK):
    for word in phrase:
            stemmedWord = stemmer.stem(word)
            resultStemming[x].append(stemmedWord)
```

```python
# Resultado do stemming
for i in range(len(resultStemming)):
    print("Frase "+str(i+1)+ " Stemming : " + str(resultStemming[i]))
```

#**spaCy**

# Lematização 
Lematizar reduz as palavras a sua forma base, mas diferente do stemming leva em conta o contexto da palavra


```python
# Baixando os dados da biblioteca
spacy.cli.download('pt_core_news_lg')
```

```python
# Carregando o modelo nessa variavel
nlp = spacy.load("pt_core_news_lg")
```

```python
# Lematização das frases 
lemmatizationSpaCy = []
for x in texts:
    doc = nlp(x)
    lemmas = [token.lemma_ for token in doc if token.pos_ == 'VERB']
    lemmatizationSpaCy.append(lemmas)

lemmatizationSpaCy
```

# Part of Speech 
Determinando a classe gramatical de cada palavra na frase 

```python
spacy.cli.download('pt_core_news_lg')
```

```python
nlp = spacy.load("pt_core_news_lg")
```

```python
# Importando o pandas para facilitar a visualização dos dados 
import pandas as pd
df = pd.DataFrame(columns=['Palavra', 'Classe Gramatical', 'Termo sintático'])
```

```python
# POS no texto de input
for x in texts:
    doc = nlp(x)
    for token in doc:
    # Selecionar o texto, a classe gramatical e o termo sintático de um token
        token_text = token.text
        token_pos = token.pos_
        token_dep = token.dep_
        df = df.append({'Palavra':token_text, 'Classe Gramatical': token_pos, 'Termo sintático': token_dep}, ignore_index=True)


```

```python
# Resultado do POS no texto 
df
```

# Reconhecimento de entidades 
Identifica e classifica as palavras 
no texto como cidades por exemplo 

```python
nlp = spacy.load("pt_core_news_lg")
```

```python
# Criando um data frame para facilitar a visualização dos dados 
dfNER = pd.DataFrame(columns=['Palavra', 'Classificação','Frase'])
```

```python
#NER no texto, foi necessário fazer um loop dentro de outro pois o programa não conseguiu lidar com a lista simples

for x in texts:
    doc = nlp(x)
    for ent in doc.ents:
        # Adicionando no data frame
        dfNER = dfNER.append({'Palavra':ent.text, 'Classificação': ent.label_, 'Frase': doc}, ignore_index=True)

```

```python
# Resultado da iteração
dfNER
```
